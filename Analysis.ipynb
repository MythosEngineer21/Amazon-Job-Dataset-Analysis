# Amazon Job Dataset

### Job Location

```python
import csv
import matplotlib.pyplot as plt

# Open the CSV file with the correct encoding
with open("amazon_jobs_dataset.csv", "r", encoding="utf-8") as f:
    f1 = csv.DictReader(f, skipinitialspace=True)
    w = {}

    # Count job openings in specific locations
    for r in f1:
        k = r['location']
        v = 1
        if k == 'IN, KA, Bangalore ' or k == 'US, WA, Seattle ':
            if k in w:
                w[k] += 1
            else:
                w[k] = 1

# Print and visualize the data
print(w['IN, KA, Bangalore '], w['US, WA, Seattle '])

# Prepare data for plotting
locations = ['Bangalore', 'Seattle']
openings = [w['IN, KA, Bangalore '], w['US, WA, Seattle ']]

# Create a bar chart
plt.bar(locations, openings)
plt.xlabel('Locations')
plt.ylabel('Number of Job Openings')
plt.title('Number of Job Openings in Bangalore and Seattle')
plt.show()
```

### Job Openings by Month in 2018

```python
import csv
import matplotlib.pyplot as plt

with open('amazon_jobs_dataset.csv', encoding='utf-8') as file_obj:
    file_data = csv.DictReader(file_obj, skipinitialspace=True)
    file_list = list(file_data)

dict = {}
month = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

for row in file_list[0:]:
    if '2018' in row['Posting_date']:
        for m in month:
            if m in row['Posting_date']:
                dict[m] = dict.get(m, 0) + 1

count = max(dict, key=dict.get)
for k, v in dict.items():
    if count == k:
        print(k, v)

# Prepare data for plotting
months = list(dict.keys())
job_counts = list(dict.values())

# Create a bar chart
plt.bar(months, job_counts)
plt.xlabel('Months')
plt.ylabel('Number of Job Openings')
plt.title('Number of Job Openings in Each Month of 2018')
plt.xticks(rotation=45)
plt.show()
```

### Job Openings Requiring Programming Languages

```python
import csv
import matplotlib.pyplot as plt

with open("amazon_jobs_dataset.csv",encoding='utf-8') as file_obj:
    file_data = csv.DictReader(file_obj, skipinitialspace=True)
    file_list = list(file_data)

dict = {}
language = ["Java", "C++", "Python"]

for row in file_list[0:]:
    if 'Bachelor' in row['BASIC QUALIFICATIONS'] or 'BS' in row['BASIC QUALIFICATIONS'] or 'BA' in row['BASIC QUALIFICATIONS']:
        for m in language:
            if 'IN' in row['location'][:2]:
                if m in row['BASIC QUALIFICATIONS']:
                    dict[m] = dict.get(m, 0) + 1

count = max(dict, key=dict.get)
for k, v in dict.items():
    if count == k:
        print(k, v)

# Prepare data for plotting
languages = list(dict.keys())
job_counts = list(dict.values())

# Create a pie chart
plt.pie(job_counts, labels=languages, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Percentage of Job Openings Requiring Programming Languages in India')
plt.show()
```

### Job Openings Requiring Java Skills by Country

```python
import csv
import matplotlib.pyplot as plt

with open("amazon_jobs_dataset.csv",encoding='utf-8') as file_obj:
    file_data = csv.DictReader(file_obj, skipinitialspace=True)
    file_list = list(file_data)

country_jobs = []
for row in file_list[0:]:
    if 'Java' in row['BASIC QUALIFICATIONS']:
        country_jobs.append(row)

d = {}
for row in country_jobs:
    d[row['location'][:2]] = d.get(row['location'][:2], 0) + 1

max_jobs = max(d.values())
max_country = [k for k, v in d.items() if v == max_jobs][0]

print(f"Country with the most job openings requiring Java skills: {max_country} ({max_jobs} job openings)")

# Prepare data for plotting
countries = list(d.keys())
job_counts = list(d.values())

# Create a bar chart
plt.bar(countries, job_counts)
plt.xlabel('Country')
plt.ylabel('Number of Job Openings')
plt.title('Number of Job Openings Requiring Java Skills by Country')
plt.xticks(rotation=45)
plt.show()
```
